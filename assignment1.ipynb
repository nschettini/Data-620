```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
node_names = (['Andre','Beverly','Carol','Diane','Ed','Fernando','Garth','Heather','Ike','Jane'])
G.add_nodes_from(node_names)
```

```python
edges = [('Andre', 'Beverly'),
         ('Andre', 'Carol'),
         ('Andre', 'Diane'),
         ('Andre', 'Fernando'),
         ('Beverly', 'Diane'),
         ('Beverly', 'Garth'),
         ('Beverly', 'Ed'),
         ('Carol', 'Fernando'),
         ('Carol', 'Diane'),
         ('Ed', 'Diane'),
         ('Ed', 'Garth'),
         ('Diane', 'Fernando'),
         ('Diane', 'Garth'),
         ('Fernando','Garth'),
         ('Heather', 'Fernando'),
         ('Heather', 'Garth'),
         ('Heather', 'Ike'),
         ('Ike','Jane' )
        ]
G.add_edges_from(edges)
```

```python
plt.axis('off')
p = nx.spring_layout(G)
nx.draw_networkx_nodes(G,pos=p,node_size=500, alpha=.5,node_color="b", node_shape="s", linewidths=3)
nx.draw_networkx_edges(G,pos=p,edge_color="c", width=2)
nx.draw_networkx_labels(G, pos=p, font_size=12, font_color="c" ,font_weight="bold")
plt.show()
```
